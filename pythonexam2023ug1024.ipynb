```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("hello world")
```

```python
#First five elements are
data=pd.read_csv("AQI_Data.csv")
print("First five elements are")
data.head(5)
```

```python
# Last six elements are
print("Last six elements are")
data.tail(6)
```

```python
# Summary of all columns
print("Summary of all columns")
data.describe()
```

```python
#mean of aqi, pm2.5, pm10 for each cities using numpy
cities = data['City'].values
aqi = data['AQI'].values
pm25 = data['PM2.5'].values
pm10 = data['PM10'].values


unique_cities = np.unique(cities)


for city in unique_cities:
    city_mask = cities == city
    mean_aqi = np.mean(aqi[city_mask])
    mean_pm25 = np.mean(pm25[city_mask])
    mean_pm10 = np.mean(pm10[city_mask])
    
    print(f"City: {city}, Mean AQI: {mean_aqi}, Mean PM2.5: {mean_pm25}, Mean PM10: {mean_pm10}")
```

```python
# Count exceeding 100
filtered_data = data[data['PM2.5'] > 100]


count_exceeding_pm25 = filtered_data.groupby('City').size()
print("Count exceeding 100")
print(count_exceeding_pm25)
```
